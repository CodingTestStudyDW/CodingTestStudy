## 백준(1260번/실버2/DFS, BFS) : DFS와 BFS / [문제 링크](https://www.acmicpc.net/problem/1260)

```python
## Version 1 (줄은 좀더 짧지만 좀더 비효율적임)
from collections import deque

# dfs 메서드 정의
def dfs(graph, v, visited, result=[]):
    # 현재 노드 방문 처리
    visited[v] = True
    result.append(v)
    # 현재 노드와 연결된 다른 노드를 재귀적으로 방문
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited, result)

    return result

# BFS메서드 정의 (재귀함수는 사용하지 않음)
def bfs(graph, start, visited, result=[]):
    # Queue구현을 위해 deque 라이브러리 사용
    queue = deque([start])
    # 현재 노드 방문 처리
    visited[start] = True
    
    # 큐가 empty상태가 될때까지 반복
    while queue:
        # 큐에서 가장 먼저들어온 하나의 원소를 뽑아 출력
        v = queue.popleft()
        result.append(v)
        # 아직 방문하지 않은 인접한 원소들을 큐에 삽입
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True

    return result

N, M, V = map(int, input().split())

graph = [[] for _ in range(N+1)]
for _ in range(M):
    a,b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)
    # 숫자가 작은쪽을 먼저 가도록 유도하기 위해서 sort진행
    
    ############################
    graph[a].sort()
    graph[b].sort()

    set(graph[a])
    set(graph[b])
    ############################

dfs_visited = [False]*(N+1)
dfs_result = dfs(graph, V, dfs_visited)
bfs_visited = [False]*(N+1)
bfs_result = bfs(graph, V, bfs_visited)

print(*dfs_result)
print(*bfs_result)
```
